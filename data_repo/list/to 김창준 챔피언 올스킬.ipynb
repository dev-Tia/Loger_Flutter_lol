```python
import requests
import time
```

# 최신버전 확인

```python
url = 'https://ddragon.leagueoflegends.com/api/versions.json'
versions = requests.get(url).json()
last_version = versions[0]

print("last_version : " + last_version)
```

# Realms 정보 확인

```python
url = 'https://ddragon.leagueoflegends.com/realms/kr.json'
realms = requests.get(url).json()
print(realms["cdn"])
print(realms["l"])
```

# champion 일반정보 확인

```python
url = realms["cdn"] + "/" +last_version + "/data/" + realms["l"] + "/champion.json"
champion_info = requests.get(url).json()
champion_info = champion_info["data"]
champion_info
```

# champion 이름만 뽑아서 list 형태로...

```python
champion_name_list = []

for i in champion_info:
    champion_name_list.append(i)
    
print(champion_name_list)
print(len(champion_name_list))
```

# 일반정보 데이터 정리

```python
for i in champion_name_list:
    champion_info[i]["image"] = champion_info[i]["image"]["full"]
    del champion_info[i]["stats"]
    del champion_info[i]["version"]
```

###### champion_info 로 일반정보 사용하면 됨

```python
champion_info
```

# 우선 champion 하나만 뽑아보자

```python
champion_name = champion_name_list[45]
url = realms["cdn"] + "/" +last_version + "/data/" + realms["l"] + "/champion/" + champion_name + ".json"
champion_data = requests.get(url).json()

champion_temp = champion_data["data"][champion_name]
```

```python
champion_temp
```

```python
def arrangement_spells(spell):
    for i in spell:
        print("-------------------------------------------------------")
        del i["tooltip"]
        if "leveltip" in i:
            del i["leveltip"]
        del i["effect"]
        del i["effectBurn"]
        i["image"] = i["image"]["full"]
        print(i)
    return spell
```

```python
champion_temp["image"] = champion_temp["image"]["full"]
del champion_temp["skins"]
del champion_temp["blurb"]
del champion_temp["stats"]

champion_temp["spells"] = arrangement_spells(champion_temp["spells"])

champion_temp["passive"]["image"] = champion_temp["passive"]["image"]["full"]

del champion_temp["recommended"]
```

###### champion_temp 는 이런식으로 생김

```python
champion_temp
```

# 이제 전체다 돌리자!!

```python
def arrangement_spells_2(spell):
    for i in spell:
        del i["tooltip"]
        
        if "leveltip" in i:
            del i["leveltip"]

        del i["effect"]
        del i["effectBurn"]
        i["image"] = i["image"]["full"]
    return spell
```

```python
all_champion_data ={}

count = 0

for i in champion_name_list:
    url = realms["cdn"] + "/" +last_version + "/data/" + realms["l"] + "/champion/" + i + ".json"
    champion_data = requests.get(url).json()
    print(str(count) + i)
    time.sleep(1)

    champion_temp = champion_data["data"][i]
    # ----------------------------------------------------------
    champion_temp["image"] = champion_temp["image"]["full"]
    del champion_temp["skins"]
    del champion_temp["blurb"]
    del champion_temp["stats"]

    champion_temp["spells"] = arrangement_spells_2(champion_temp["spells"])

    champion_temp["passive"]["image"] = champion_temp["passive"]["image"]["full"]

    del champion_temp["recommended"]
    # ----------------------------------------------------------    
    all_champion_data[i] = champion_temp
    
    count += 1
```

###### champion_temp 는 이런식으로 생김 (214KB)

```python
all_champion_data
```

```python
check_list = []

for i in all_champion_data:
    check_list.append(i)
    
print(check_list)
print(len(check_list))
```
